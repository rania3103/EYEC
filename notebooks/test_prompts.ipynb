#### In this jupyter notebook we will test many prompts and chose the optimized one for meme

```python
# import modules
import matplotlib.pyplot as plt
from PIL import Image
import sys
import os
# add project root
sys.path.append(os.path.abspath(os.path.join("..")))
import time
```

```python
# show img
img_path_ = "../data/images/image_1.jpg"
img = Image.open(img_path_)
plt.imshow(img)
plt.title("Test image")
plt.axis("off")
plt.show()
```

#### testing prompts

```python
# import OvisCaptioner class
from app.app import OvisCaptioner

captioner = OvisCaptioner()
prompts = [
    "Describe the image.",
    "You are a vision assistant for blind people. First detect meme, then caption.",
    "Is this a meme? If yes, humorously caption.",
    "You are a vision-language assistant for blind users. Extract memes or celebrity names if present, else describe the scene.",
    "A vivid, emotionally aware caption for a blind person describing this image scene, any humor, any famous people, and any overlaid text naturally as part of the description.",
    "Describe this image vividly for a blind person. If it’s a meme, keep it humorous. If any text appears, mention it. If famous people or characters appear, mention them naturally.",
    "You are a helpful assistant for blind people. Describe the scene, identify any characters or celebrities, read any overlaid text, and mention if it's a meme.",
    "Provide a descriptive caption for this image including any text and identifying any famous characters, memes, or humor present.",
    "SYSTEM: Always describe images for blind people with vivid emotions, identify text, humor, and celebrity names when present. Avoid saying 'this is a meme'; just caption it humorously if so."
    "Caption this image quickly. Is it a meme? A celebrity? Just state what's happening without extra explanation.", 
    "Meme or person or what exactly? Output only the description, no fluff.",
    "Quickly describe this image. If there's a face, name the person. If it's a meme, summarize the joke.",
    "You are a vision assistant for blind people this is a meme caption it.",
    "humorously caption this meme."
]
for prompt in prompts:
    start_time = time.time()
    result = captioner.describe_image(img_path=img_path_, text=prompt)["caption"]
    end_time = time.time()
    inf_time = end_time - start_time
    print(f"Prompt: {prompt}")
    print(f"Inference time: {inf_time:.2f}s")
    print(f"Caption: {result}")
    print("="*60)
```

we will choose this prompt fot its low inference time and high quality description extracting text and reasoning:
