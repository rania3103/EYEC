```python
import asyncio
from PIL import Image
from edge_tts import Communicate
from IPython.display import Audio, display, Image as IPImage
import nest_asyncio
import os
import sys
import matplotlib.pyplot as plt
# add project root
sys.path.append(os.path.abspath(os.path.join("..")))
# test
caption_text = "hello"
nest_asyncio.apply()
# TTS
async def generate_tts(text, output_path="caption_audio.mp3"):
    communicate = Communicate(text, "en-US-JennyNeural")
    await communicate.save(output_path)

# run TTS
asyncio.run(generate_tts(caption_text, "../tts/caption_audio.mp3"))

# play audio
display(Audio("../tts/caption_audio.mp3"))
```

```python
# show img
img_path1 = "../data/images/image_1.jpg"
img = Image.open(img_path1)
plt.imshow(img)
plt.title("Test image")
plt.axis("off")
plt.show()
```

```python
from app.app import OvisCaptioner

captioner = OvisCaptioner()
```

```python
result = captioner.describe_image(img_path=img_path1)
caption = result["caption"]

print("Caption:", caption)

# TTS
asyncio.run(generate_tts(caption, "../demo_readme/caption_audio1.mp3"))
display(Audio("../demo_readme/caption_audio1.mp3"))
```
